```python
df['Classification'].value_counts().plot(kind='pie')
```

```python
df[['SL','PL']].value_counts().head(15).plot(kind='bar')
plt.title('Sepal and petal length',fontsize=18)
plt.xlabel('SL',fontsize=15)
plt.ylabel('PL',fontsize=15)
plt.show
```

```python
df.groupby('Classification')['SW'].sum().plot(kind="bar")
```

```python
sns.lineplot(data=df,x='SW',y='PW')
plt.title('Line chart of Sepal width over Petal Width')
plt.xlabel('SW')
plt.ylabel('PW')
```
